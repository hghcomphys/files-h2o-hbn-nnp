```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

## Cutoff fuction

```python
def fcut(r, rcut):
    
    if (r > rcut): 
        return 0.0

    # COS TYPE
    return ( np.cos(np.pi * r /rcut) + 1.0 ) * 0.5; 
    
    # TANH TYPE
#     tmp = np.tanh(1.0 - r / rcut);
#     return tmp * tmp * tmp;
```

## Symmetry function (radial)

```python
def G1(r, rcut, eta=0.0, rshift=0.0):
    return np.exp( -eta*(r-rshift)**2 ) * fcut(r, rcut)
```

```python
def plot_G1(rcut, rshift_list, eta_list, elements):
    
    r = np.linspace(0+1e-10, rcut, 1000)
    for rs in rshift_list:
        for et in eta_list:
            plt.plot(r, [G1(_, rcut, et, rs) for _ in r], label="eta=%4.4f, rs=%2.2f"%(et, rs))
            print("symfunction_short", elements[0], 2, elements[1], " %6.4f %5.2f %5.2f"%(et, rs, rcut))

    plt.axvline(rmin, color='k', linestyle='--', label='r_min')
    plt.legend()
    plt.xlabel('distance')
    plt.ylabel('radial symmetry function')
    
```

## Symmetry functions (angular)

```python
def G3(rcut, zeta, lamb, eta, elements):
    
    n_elem = len(elements)
    nsf = 0
    
    for ele1 in elements:     
        for i2 in range(n_elem):
            ele2 = elements[i2]
            for i3 in range(n_elem):
                ele3 = elements[i3]
                
                if i2 < i3:
                    continue
                
                for et in eta:
                    for zt in zeta:
                        for lm in lamb:
                            print ("symfunction_short", ele1, 3, ele2,  ele3,  " %6.4f %5.2f %5.2f %5.2f"%(et, lm, zt, rcut))
                            nsf+=1
        print()
    print(nsf)
```
